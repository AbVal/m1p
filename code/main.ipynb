# Эксперименты с RandomForestClassifier

## steel plates

```python
dataset = prepare_steel_plates()

search_space = RF_search_space()
model1 = DEHB(f=RFC_target_function, 
             cs=search_space, 
             min_budget=2, 
             max_budget=20,
             n_workers=1,
             output_path="./rand1_bin_logs",
             strategy='rand1_bin')
model2 = DEHB(f=RFC_target_function, 
             cs=search_space, 
             min_budget=2, 
             max_budget=20,
             n_workers=1,
             output_path="./noisy1_bin_logs",
             strategy='noisy1_bin',
             basic_noise_scale=0.0,
             noise_step=0.15)

run1_sp, avg_test_score_1_sp, run2_sp, avg_test_score_2_sp, fig = run_comparison_experiment(model1, model2, dataset, fevals=20, runs=20,
                              fig_title='Зависимость ошибки на валидации от количества итераций DEHB',
                              fig_name='steel_plates_cmp.pdf')
print('DEHB average test score:', avg_test_score_1_sp)
print('N-DEHB average test score:', avg_test_score_2_sp)
```
